```python


subset = ['Facility ID',
 'Facility Name',
 'Address',
 'City/Town',
 'State',
 'ZIP Code',
 'County/Parish',
 'Telephone Number',
 'Hospital Type',
 'Hospital Ownership',
 'Emergency Services',
 'Meets criteria for birthing friendly designation',
 'Hospital overall rating']

df_subset = pd.DataFrame(df_la[subset])
df_subset.reset_index(drop=True, inplace=True)
```

```python
df_subset.info()
```

```python
df_subset.head()
```

```python
print({column:'___' for column in df_subset.columns})
```

```python
df_subset.rename(
    columns={'Facility ID': 'fac_id',
             'Facility Name': 'fac_name',
             'Address': 'address',
             'City/Town': 'city',
             'State': 'state',
             'ZIP Code': 'zip_code',
             'County/Parish': 'county',
             'Telephone Number': 'tel_num',
             'Hospital Type': 'hosp_type',
             'Hospital Ownership': 'hosp_owner',
             'Emergency Services': 'emergency',
             'Meets criteria for birthing friendly designation': 'maternity',
             'Hospital overall rating': 'rating'},
    inplace=True
)
```

```python
df_subset['emergency'] = np.where(df_subset['emergency'] == "Yes", 1, 0)
df_subset['maternity'] = np.where(df_subset['maternity'] == "Y", 1, 0)
```

```python
df_subset
```

```python
df_subset['urgent_care'] = np.where(df_subset['hosp_type'] == 'Acute Care Hospitals', 1, 0)
df_subset['psychiatric'] = np.where(df_subset['hosp_type'] == 'Psychiatric', 1, 0)
df_subset['childrens'] = np.where(df_subset['hosp_type'] == 'Childrens', 1, 0)
df_subset['veterans'] = np.where(df_subset['hosp_owner'] == 'Veterans Health Administration', 1, 0)
df_subset
```
