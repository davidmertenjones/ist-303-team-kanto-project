```python
df_la['Hospital overall rating'] = df_la['Hospital overall rating'].apply(lambda x: x.replace('Not Available', '-1'))
df_la['Hospital overall rating'] = df_la['Hospital overall rating'].apply(lambda x: int(x))
```

```python
import matplotlib.pyplot as plt
plt.hist(df_la[df_la['Hospital overall rating'] > 0]['Hospital overall rating'],
        bins=9)
plt.title('Distribution of Hospital Overall Ratings')
plt.show()
```
